```python
# prompt: use getpass to save username and password into variables

# import getpass

# set up sftp access
# username and password are stored within Colab secrets
hostname = "mxre-e2e-files.parentlink.net"
```

```python
# prompt:  connect to an sftp and list the files

from google.colab import userdata # this enables the use of secrets
import paramiko

try:
    transport = paramiko.Transport((hostname, 22))
    # transport.connect(username=username, password=password)
        # this now leverages Colab notebook secrets
    transport.connect(username=userdata.get('mxre-e2e-sftp_user'), password=userdata.get('mxre-e2e-sftp_pw'))

    sftp = paramiko.SFTPClient.from_transport(transport)

    files = sftp.listdir()
    print("Files in the remote directory:")
    for file in files:
        print(file)

    sftp.close()
    transport.close()

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# load a file into a pandas df and display the head

import pandas as pd

try:
    transport = paramiko.Transport((hostname, 22))
    transport.connect(username=userdata.get('mxre-e2e-sftp_user'), password=userdata.get('mxre-e2e-sftp_pw'))

    sftp = paramiko.SFTPClient.from_transport(transport)
    filename = 'CustomParentsBig Spring ISDData_anonimized.txt'

    # Download the file
    remote_file_path = filename  # Replace with the actual file path on the server
    local_file_path = "/tmp/test.txt" # Specify the local path to save the file.
    sftp.get(remote_file_path, local_file_path)

    # Load the data into a pandas DataFrame
    try:
      df = pd.read_csv(local_file_path, delimiter='\t') # Assumes tab-separated file. Change delimiter if needed
    except pd.errors.ParserError:
        print("Error: Could not parse the file. Check the delimiter or file format.")

    sftp.close()
    transport.close()

except Exception as e:
    print(f"An error occurred: {e}")

print("Reading from the "+filename+" file on the SFTP")
df.head()
```

```python
# prompt: find items in df where parentid like "2b26f0e205d55b2fb237f42e5c918a21ff767def"

# Convert the parentid column to string type to avoid potential errors
df['parentid'] = df['parentid'].astype(str)

# Use the .str.contains() method for partial string matching
matching_rows = df[df['parentid'].str.contains("2b26f0e205d55b2fb237f42e5c918a21ff767def", na=False)]

# Print the matching rows
matching_rows
```
